```python
# Caminho dos arquivos
arquivo_mun = r'\\wsl$\Ubuntu\home\edcarlos\work_environment\datalake\bronze\exportacao_mun_2024.csv'
arquivo_ncm_sh = r'\\wsl$\Ubuntu\home\edcarlos\work_environment\datalake\landing\NCM_SH.csv'

# Carregar o arquivo principal que contém a coluna cd_pais
df = pd.read_csv(arquivo_mun, sep=',', encoding='latin1')

# Carregar o arquivo que contém os códigos e nomes dos países
df_ncm_sh = pd.read_csv(arquivo_ncm_sh, sep=';', encoding='latin1')

# Garantir que cd_sh4 seja string
df['cd_sh4'] = df['cd_sh4'].astype(str)

# Garantir que CO_SH4 seja string
df_ncm_sh['CO_SH4'] = df_ncm_sh['CO_SH4'].astype(str)

# Fazer o merge para substituir os códigos pelos nomes dos países
df = df.merge(df_ncm_sh, left_on='cd_sh4', right_on='CO_SH4', how='left')

# Exibir as primeiras linhas para verificar se deu certo
df.head()
```
