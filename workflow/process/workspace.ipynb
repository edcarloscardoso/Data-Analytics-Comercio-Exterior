```python
import pandas as pd
# Caminho do arquivo
caminho_arquivo = r'\\wsl$\Ubuntu\home\edcarlos\Data-Analytics-Comercio-Exterior\datalake\raw_data\EXP_2023.csv' #foi nomeado dessa forma porque não encontrou o arquivo porque está no wsl2
# Lendo o CSV para um DataFrame
df = pd.read_csv(caminho_arquivo, sep=';')
# Exibir as primeiras linhas do DataFrame
print(df.head())
```

```python
#renomear coluna
df = df.rename(columns={
    'CO_ANO': 'dt_ano',
    'CO_MES': 'dt_mes',
    'CO_NCM': 'cd_ncm',
    'CO_UNID': 'cd_unidade',
    'CO_PAIS': 'cd_pais',
    'SG_UF_NCM': 'sg_ufncm',
    'CO_VIA': 'cd_via',
    'CO_URF': 'cd_urf',
    'QT_ESTAT': 'qt_estatistica',
    'KG_LIQUIDO': 'kg_liquido',
    'VL_FOB': 'vl_dolarfob'
    })
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.head() #verificar colunas
```

```python
df['data_exportacao'] = pd.to_datetime(df['dt_ano'].astype(str) + '-' + df['dt_mes'].astype(str)) #unir as colunas ano e mes e transformar em data
```

```python
df.head()
```

```python
df['sg_ufncm'] = df['sg_ufncm'].str.strip().str.lower() #colocar as siglas dos Estados em minúsculo
```

```python
df.head()
```

```python
df = df.drop(columns=['dt_ano', 'dt_mes']) #remover coluna
```

```python
df.head()
```

```python
caminho = "/home/edcarlos/Data-Analytics-Comercio-Exterior/datalake/landing/exportacao_2023.csv" #salvando o arquivo
df.to_csv(caminho, index=False)
print("Arquivo salvo com sucesso!")
```
