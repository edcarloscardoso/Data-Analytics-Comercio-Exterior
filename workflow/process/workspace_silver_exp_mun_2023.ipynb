```python
# Carregar o arquivo principal que contém a coluna cd_ncm
df = pd.read_csv(arquivo_imp_mun_2023, sep=',', encoding='latin1')

# Carregar o arquivo que contém a coluna CO_NCM
df_ncm = pd.read_csv(arquivo_ncm, sep=';', encoding='latin1')

# Garantir que cd_ncm seja string
df['cd_sh4'] = df['cd_sh4'].astype(str)

# Garantir que CO_NCM seja string
df_ncm['CO_SH4'] = df_ncm['CO_SH4'].astype(str)

# Fazer o merge para substituir os códigos pelos nomes dos países
df = df.merge(df_ncm, left_on='cd_sh4', right_on='CO_SH4', how='left')

# Exibir as primeiras linhas para verificar se deu certo
df.head()
```
