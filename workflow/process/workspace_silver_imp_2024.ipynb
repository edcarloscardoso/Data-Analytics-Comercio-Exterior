```python
def carregar_e_mesclar_arquivos(url_imp_2024: str, url_ncm: str, 
                                chave_principal: str, chave_referencia: str, 
                                sep_principal: str = ',', sep_referencia: str = ';', 
                                encoding: str = 'latin1'):
    
    # Carregar os arquivos dentro da função
    df_imp_2024 = pd.read_csv(url_imp_2024, sep=sep_principal, encoding=encoding)
    df_ncm = pd.read_csv(url_ncm, sep=sep_referencia, encoding=encoding)

    # Garantir que as colunas de chave sejam strings
    df_imp_2024[chave_principal] = df_imp_2024[chave_principal].astype(str)
    df_ncm[chave_referencia] = df_ncm[chave_referencia].astype(str)

    # Realizar o merge
    df_resultado = df_imp_2024.merge(df_ncm, left_on=chave_principal, right_on=chave_referencia, how='left')

    return df_resultado

# Exemplo de uso
df_1 = carregar_e_mesclar_arquivos(
    url_imp_2024=r'\\wsl$\Ubuntu\home\edcarlos\Data-Analytics-Comercio-Exterior\datalake\landing\importacao_2024.csv',
    url_ncm=r'\\wsl$\Ubuntu\home\edcarlos\Data-Analytics-Comercio-Exterior\datalake\raw_data\NCM.csv',
    chave_principal='cd_ncm',
    chave_referencia='CO_NCM'
)

# Exibir as primeiras linhas para verificar
df_1.head()
```

```python
# Caminho do arquivo
url_urf = r'\\wsl$\Ubuntu\home\edcarlos\Data-Analytics-Comercio-Exterior\datalake\raw_data\URF.csv' #foi nomeado dessa forma porque não encontrou o arquivo porque está no wsl2
# Lendo o CSV para um DataFrame
df_urf = pd.read_csv(url_urf, sep=';',  encoding='latin1')  # Ou tente 'ISO-8859-1'
# Exibir as primeiras linhas do DataFrame
df_urf.head()
```

```python
# Função para mesclar os DataFrames já carregados
def mesclar_dataframes(df_base: pd.DataFrame, df_referencia: pd.DataFrame, 
                       chave_base: str, chave_referencia: str) -> pd.DataFrame:
    """
    Realiza a junção (merge) entre a base principal e o DataFrame de referência.
    """
    # Garantir que as colunas usadas no merge sejam strings
    df_base[chave_base] = df_base[chave_base].astype(str)
    df_referencia[chave_referencia] = df_referencia[chave_referencia].astype(str)

    # Realizar o merge
    df_resultante = df_base.merge(df_referencia, left_on=chave_base, right_on=chave_referencia, how='left')
    
    return df_resultante

# Exemplo de aplicação da função

# Já carregados anteriormente
# df_1 (base principal) e df_urf (referência) já estão definidos

# Mesclar os DataFrames
df_1 = mesclar_dataframes(df_base=df_1, df_referencia=df_urf, chave_base='cd_urf', chave_referencia='CO_URF')

# Exibir as primeiras linhas do DataFrame resultante
df_1.head()
```

```python
# Etapa 1: Extrair o código da coluna NO_URF
df_1['urf_codigo'] = df_1['NO_URF'].str.extract(r'^(\d+)\s*-\s*')

# Etapa 2: Remover o código e deixar só as informações
df_1['urf_info'] = df_1['NO_URF'].str.replace(r'^\d+\s*-\s*', '', regex=True)

# Etapa 3: Separar tipo e nome, se existir
def separar_info(valor):
    partes = valor.split(' - ')
    if len(partes) == 2:
        tipo, nome = partes
    else:
        tipo = None
        nome = valor
    return pd.Series([tipo, nome])

df_1[['urf_tipo', 'urf_nome']] = df_1['urf_info'].apply(separar_info)

# Etapa 4: Limpar colunas temporárias
df_1.drop(columns=['NO_URF', 'urf_info'], inplace=True)
df_1.head()
```

```python
df_1 = df_1.drop(['cd_urf', 'CO_URF', 'urf_codigo', 'urf_tipo'], axis=1) #removendo as colunas com códigos
df_1.head()
```

```python
def padronizar_colunas(df, colunas):
    """
    Remove espaços extras e converte os valores das colunas especificadas para minúsculas.

    Parâmetros:
        df (pd.DataFrame): DataFrame de entrada.
        colunas (list): Lista de colunas a serem padronizadas.

    Retorna:
        pd.DataFrame: DataFrame com as colunas padronizadas.
    """
    df[colunas] = df[colunas].apply(lambda x: x.str.strip().str.lower())
    return df

df_4 = padronizar_colunas(df_4, ['cd_ncm', 'cd_urf', 'cd_via', 'cd_pais', 'cd_unidade'])
df_4.head()
```
